```python
from ultralytics import YOLO
import cv2
import os

# Model load کریں
model = YOLO("/content/last.pt")

# Prediction چلائیں اور save کریں
results = model.predict(
    source="/content/My-First-Project-4/train/images/DSC_1008_JPG.rf.b3f2e8e2fcafa2afaf2e6752d1e1bbd3.jpg",  # folder with images
    conf=0.5,
    save=True
)

# Prediction results directory
pred_dir = results[0].save_dir
pred_images = [os.path.join(pred_dir, f) for f in os.listdir(pred_dir) if f.endswith(('.jpg', '.png'))]
```

```python
from jupyter_dash import JupyterDash
from dash import html, dcc
import base64

# Function to encode image
def encode_image(image_path):
    with open(image_path, "rb") as f:
        encoded = base64.b64encode(f.read()).decode()
    return "data:image/jpeg;base64," + encoded

# Encode prediction images
encoded_images = [encode_image(img) for img in pred_images]

# Dashboard App
app = JupyterDash(__name__)

app.layout = html.Div([
    html.H1("YOLO Object Detection Results", style={"textAlign": "center", "color": "#2E86C1"}),
    html.Div([
        html.Div([
            html.Img(src=img, style={"width": "100%", "border": "3px solid #ccc", "border-radius": "10px"})
        ], style={"width": "30%", "display": "inline-block", "margin": "1%"})
        for img in encoded_images
    ], style={"textAlign": "center"})
])

app.run(mode="inline")
```

```python
!pip install ultralytics opencv-python

from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow

# Load your trained YOLOv9 model
model = YOLO("/content/last.pt")  # change to your weights

# Open webcam (in Colab, use 0 for default cam — but will refresh frame-by-frame)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run detection
    results = model.predict(frame, conf=0.5)

    # Annotate frame
    annotated_frame = results[0].plot()

    # Show in Colab cell output
    cv2_imshow(annotated_frame)

    # Exit after first frame (Colab can't stream like a local PC)
    break

cap.release()
```
